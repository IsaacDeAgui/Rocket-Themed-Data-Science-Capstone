Next, let's create markers for all launch records.
If a launch was successful `(class=1)`, then we use a green marker and if a launch was failed, we use a red marker `(class=0)`


Note that a launch only happens in one of the four launch sites, which means many launch records will have the exact same coordinate. Marker clusters can be a good way to simplify a map containing many markers having the same coordinate.


Let's first create a `MarkerCluster` object


```python
marker_cluster = MarkerCluster()
```

*TODO:* Create a new column in `launch_sites` dataframe called `marker_color` to store the marker colors based on the `class` value


```python
from folium.plugins import MarkerCluster

# Create a MarkerCluster object
marker_cluster = MarkerCluster()

# Create a new column in launch_sites dataframe called marker_color
launch_sites['marker_color'] = 'blue'  # Default color for sites without launches

# Update marker colors based on the class value from spacex_df
for index, site in launch_sites.iterrows():
    site_outcomes = spacex_df[spacex_df['Launch Site'] == site['Launch Site']]
    
    # Check if there are any successful launches for the site
    if any(site_outcomes['class'] == 1):
        launch_sites.at[index, 'marker_color'] = 'green'  # Set color to green for successful launches
    
    # Check if there are any failed launches for the site
    elif any(site_outcomes['class'] == 0):
        launch_sites.at[index, 'marker_color'] = 'red'  # Set color to red for failed launches

# Display the updated launch_sites DataFrame
launch_sites

# Apply a function to check the value of `class` column
# If class=1, marker_color value will be green
# If class=0, marker_color value will be red
```

*TODO:* For each launch result in `spacex_df` data frame, add a `folium.Marker` to `marker_cluster`


```python
# Create a new column in the spacex_df DataFrame called marker_color
spacex_df['marker_color'] = spacex_df['class'].apply(lambda x: 'green' if x == 1 else 'red')

# Now, iterate through each launch result in spacex_df data frame and add a folium.Marker to marker_cluster
for index, record in spacex_df.iterrows():
    # Create a Marker object with its coordinate
    marker = folium.Marker(location=[record['Lat'], record['Long']])
    
    # Customize the Marker's icon property to indicate if this launch was successful or failed
    marker.icon = folium.Icon(color='white', icon_color=record['marker_color'])
    
    # Add the Marker to the MarkerCluster
    marker_cluster.add_child(marker)

# Add marker_cluster to the current site_map
site_map.add_child(marker_cluster)

# Display the site_map
site_map
```

```python
# TASK 3: Calculate the distances between a launch site to its proximities
from math import radians, sin, cos, sqrt, atan2

def haversine_distance(lat1, lon1, lat2, lon2):
    # Convert latitude and longitude from degrees to radians
    lat1, lon1, lat2, lon2 = map(radians, [lat1, lon1, lat2, lon2])

    # Haversine formula
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    # Radius of the Earth in kilometers (mean value)
    radius = 6371.0

    # Calculate the distance
    distance = radius * c

    return distance

# Example usage:
launch_site_lat = 28.562302
launch_site_lon = -80.577356

# Calculate distance to a proximity point
proximity_lat = 28.573255
proximity_lon = -80.646895

distance_to_proximity = haversine_distance(launch_site_lat, launch_site_lon, proximity_lat, proximity_lon)
print(f"Distance to proximity point: {distance_to_proximity} km")
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```
